```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
plt.figure(figsize=(19, 8))
sns.barplot(x = data.Category, y = data.Score, hue = data.Model, data = data)
plt.xticks(rotation=30)
plt.ylabel("AUC Scores")
plt.show()
```
